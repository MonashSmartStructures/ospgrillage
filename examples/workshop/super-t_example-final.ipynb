### Units
Establish a consistent set of units - in this case the base units are N and m.


```python
kilo = 1e3
milli = 1e-3
N = 1
m = 1
mm = milli * m
m2 = m ** 2
m3 = m ** 3
m4 = m ** 4
kN = kilo * N
Pa = 1
MPa = N / ((mm) ** 2)
GPa = kilo * MPa
kPa = kilo * Pa
```

### Materials

Create a `Material` object using `create_material()` function.

```python
concrete = og.create_material(material="concrete", code="AS5100-2017", grade="65MPa")
```

Alternatively, users can always create customize concrete material with:

1. Elastic modulus, `E`
2. Possion ratio, `v`
3. Density, `rho`

If shear modulus, $G$, is not defined `ospgrillage` automatically determines the shear modulus using 
$ G = \frac{E}{2 (1+v)}  $

```python
concrete_custom = og.create_material(material="concrete",E=50*GPa, v= 0.3, rho=24*kN/m3 )
```

## Create the Grillage Model

Creating model using `create_grillage()` function. Then assigning the defined `GrillageMember`s to the elements using `set_member()` function.


```python
model = og.create_grillage(
    bridge_name="Super-T 33_5m",
    long_dim=L,
    width=w,
    skew=angle,
    num_long_grid=n_l,
    num_trans_grid=n_t,
    edge_beam_dist=edge_dist,
    ext_to_int_dist = ext_to_int_dist,
)
```

In `ospgrillage`, we `set_member()` using [names for standard grillage element](https://monashsmartstructures.github.io/ospgrillage/rst/Module%20description.html#assigning-grillage-members). For the standard non-skew bridge mesh, we set the :

```python
# assign grillage member to element groups of grillage model
model.set_member(longitudinal_beam, member="interior_main_beam")
model.set_member(longitudinal_beam, member="exterior_main_beam_1")
model.set_member(longitudinal_beam, member="exterior_main_beam_2")
model.set_member(edge_longitudinal_beam, member="edge_beam")
model.set_member(transverse_slab, member="transverse_slab")
model.set_member(end_transverse_slab, member="start_edge")
model.set_member(end_transverse_slab, member="end_edge")
```

Last step, create the model in OpenSees using:

### Dead loads 

Dead loads comprise of the self-weight of Super-T girders and overlay slab

For the comparison in this example, we shall split the Super-T girder and overlay slab's load cases into individual load cases.


#### Self-weight of Super-T girders


The self-weight of super-T girders is calculated as follows:

$ w_{girder} = 24 ~ \mathrm{kN/m^3} \times 0.934 ~ \mathrm{m^2} = 22.4 ~ \mathrm{kN/m} $

To this, we create a `LineLoad` and assign along position of longitudinal beams. Note for this example we assume this weight for all beams. 

```python
start_beam = 0
end_beam = L
beam_mag = 22.4 * kN / m

# create Super T self weight load case 
DL_self_weight = og.create_load_case(name="Super T self weight")

for z_pos in model.Mesh_obj.noz[1:-1]:
    beam_load_position_1 = og.create_load_vertex(x=start_beam, z=z_pos, p=beam_mag)
    beam_load_position_2 = og.create_load_vertex(x=end_beam, z=z_pos, p=beam_mag)

    self_weight_beam = og.create_load(loadtype="line", point1=beam_load_position_1, 
                                      point2=beam_load_position_2,)

    DL_self_weight.add_load(self_weight_beam)

model.add_load_case(DL_self_weight)
```

#### Overlay slab

A 180  mm in situ concrete overlay slab forms the deck. The distributed weight (area) of the deck is calculated as:

$ w_{slab} = 0.18 ~ \mathrm{m} \times 24 ~ \mathrm{kN/m^3} = 4.32 ~ \mathrm{kN/m^2}  $

We use a `PatchLoad` to define the overlay slab weight.


```python
overlay_p_mag = 4.32 * kN / m2 
patch_vertex_1 = og.create_load_vertex(x=0, z=0, p=overlay_p_mag)
patch_vertex_2 = og.create_load_vertex(x=L, z=0, p=overlay_p_mag)
patch_vertex_3 = og.create_load_vertex(x=L, z=w, p=overlay_p_mag)
patch_vertex_4 = og.create_load_vertex(x=0, z=w, p=overlay_p_mag)
                            
overlay_slab = og.create_load(
    loadtype="patch",
    name="overlay",
    point1=patch_vertex_1,
    point2=patch_vertex_2,
    point3=patch_vertex_3,
    point4=patch_vertex_4,
)

DL_overlay = og.create_load_case(name="Overlay self weight")
DL_overlay.add_load(overlay_slab)

model.add_load_case(DL_overlay)
```

### Superimposed Dead loads (SIDL)

Superimposed dead load (SIDL) consist of asphalt surfacing and non-structural precast concrete edge barriers. 

#### Asphalt surfacing
The surfacing is 65 mm 
thick with a self-weight of 22 kN/m$^3$

The magnitude of the asphalt surfacing is calculated as follows:

$ w_{asphalt} = 0.065 ~ \mathrm{m} \times 22 ~ \mathrm{kN/m^3} = 1.43 ~ \mathrm{kN/m^2}  $


```python
# asphalt surfacing (units kN/m2) - `patch` in global coords
asphalt_udl = 1.43 * kN / m2 

patch_vertex_1 = og.create_load_vertex(x=0, z=0, p=asphalt_udl)
patch_vertex_2 = og.create_load_vertex(x=L, z=0, p=asphalt_udl)
patch_vertex_3 = og.create_load_vertex(x=L, z=w, p=asphalt_udl)
patch_vertex_4 = og.create_load_vertex(x=0, z=w, p=asphalt_udl)
                            
asphalt_surfacing = og.create_load(
    loadtype="patch",
    name="asphat surfacing",
    point1=patch_vertex_1,
    point2=patch_vertex_2,
    point3=patch_vertex_3,
    point4=patch_vertex_4,
)
```

#### Precast edge barriers

The non-structural precast concrete edge barriers weighs 6.54 kN/m.

We use a line load to represent the precast edge barriers on the edge beams.

```python
# precast edge barriers (units kN/m) - `compound` in global coords
# Compount load helps us group different types of load into a single load case
barrier_udl = 6.54 * kN/m

left_barrier_p1 = og.create_load_vertex(x=0,z=0,p= barrier_udl)
left_barrier_p2 = og.create_load_vertex(x=L,z=0,p= barrier_udl)
right_barrier_p1 = og.create_load_vertex(x=0,z=w,p= barrier_udl)
right_barrier_p2 = og.create_load_vertex(x=L,z=w,p= barrier_udl)

left_barrier = og.create_load(
    loadtype="line", name="left barrier", point1=left_barrier_p1, point2=left_barrier_p2
)
right_barrier = og.create_load(
    loadtype="line", name="right barrier", point1=right_barrier_p1, point2=right_barrier_p2
)
```

Add both Asphalt and Parapets to a load case called `SIDL`

```python
SIDL = og.create_load_case(name="SIDL")
SIDL.add_load(asphalt_surfacing)
SIDL.add_load(left_barrier)
SIDL.add_load(right_barrier)
model.add_load_case(SIDL)
```

### Australian Traffic Load Model (M1600)
This is a 12-axle moving notional vehicle with central (variable) axle spacing set to 6.25 m 
and a uniformly distributed load on each standard design lane of 3.2 m (after AS 5100.2).

```python
display.Image(b64decode(m1600_png))
```

According to AS5100, the number of design lanes on bridge is:

#### Create M1600 vehicle

For this, we use `ospgrillage`'s `CompoundLoad` object. Similar to the UDL component, we can adopt a `for` loop to generate the `PointLoad` of each axle then  add to the compound load.

```python
display.Image(b64decode(axle_top_png))
```

Here, `p1` is the 60 kN wheel load. 

With `ospgrillage`, users can write generic functions that wraps the `create_load()` function. The following example showcase how we can create the axles using a function. Note for load models, `ospgrillage` v0.1.1 has an API function `create_load_model()` for this M1600 and other types of load model. 

```python
def make_M1600_vehicle(gap):
    """
    Makes an M1600 vehicle in local coordinates
    """
    axle_dist = 1.25 * m 
    left_group_dist = 3.75 * m 
    right_group_dist = 5 * m 
    wheel_load = 60*kN
    
    # Define longitudinal positions of each axle
    load_positions_x = [
        0,
        axle_dist,
        axle_dist * 2,
        left_group_dist + axle_dist * 2 ,
        left_group_dist+ axle_dist * 2  + axle_dist,
        left_group_dist + axle_dist * 2 +  2 * axle_dist,
        gap + left_group_dist + axle_dist * 2 +  2 * axle_dist ,
        gap + left_group_dist + axle_dist * 2 +  2 * axle_dist + axle_dist,
        gap + left_group_dist + axle_dist * 2 +  2 * axle_dist + axle_dist * 2,
        right_group_dist + gap + left_group_dist + axle_dist * 2 +  2 * axle_dist + axle_dist * 2 ,
        right_group_dist + gap + left_group_dist + axle_dist * 2 +  2 * axle_dist + axle_dist * 2  + axle_dist,
        right_group_dist + gap + left_group_dist + axle_dist * 2 +  2 * axle_dist + axle_dist * 2 +  2 * axle_dist,
    ]
    # Transverse position of each wheel
    load_positions_z = [-1, 1]
    
    M1600_vehicle = og.create_compound_load(name="M1600 Vehicle")
    for z in load_positions_z:
        for x in load_positions_x:
            vert = og.create_load_vertex(x=x, z=z, p=wheel_load)
            point = og.create_load(loadtype="point",name="M1600 point", point1 = vert)
            M1600_vehicle.add_load(load_obj=point)

    return M1600_vehicle
```

According to AS5100, the accompanying lane factor factors for the three design lanes (denoted as L1, L2 and L3) are 1.0, 0.8 and 0.4 respectively.

The dynamic allowance factor of 30% is defined here.

```python
alf = [1.0, 0.8, 0.4]
dla = 1.3
```

And we'll also make a list of the elements that we will move

```python
M1600_moving_loads = []
```

Now, create the M1600 load model for each lane comprising:

- Vehicle
- UDL patch load

We create a `PatchLoad` object and M1600 vehicle on each traffic lane, and add them into a single load case using a `for` loop:

```python
gap = 6.25 * m 
m1600_L1_lc = og.create_load_case(name="M1600 L1")
m1600_L2_lc = og.create_load_case(name="M1600 L2")
m1600_L3_lc = og.create_load_case(name="M1600 L3")
m1600_lc_list = [m1600_L1_lc,m1600_L2_lc,m1600_L3_lc]

# For each lane
for i in range(3):
    # Create generic vehicle
    M1600_vehicle_generator = og.create_load_model(model_type="M1600",gap=gap)
    M1600_vehicle_n = M1600_vehicle_generator.create()
    # Position in global coords
    M1600_vehicle_n.set_global_coord(og.Point(x_coord[i], 0, z_coord[i]))
    # And add to the load case with the lane factor
    m1600_lc_list[i].add_load(load_obj=M1600_vehicle_n,load_factor=alf[i])
    # And store in list for moving load
    M1600_moving_loads.append(M1600_vehicle_n)
    
    # Now create the patch
    vertex_1 = og.create_load_vertex(x=-L, z=z_coord[i] - udl_width/2, p=udl_mag)
    vertex_2 = og.create_load_vertex(x=2*L, z=z_coord[i] - udl_width/2, p=udl_mag)
    vertex_3 = og.create_load_vertex(x=2*L, z=z_coord[i] + udl_width/2, p=udl_mag)
    vertex_4 = og.create_load_vertex(x=-L, z=z_coord[i] + udl_width/2, p=udl_mag)

    M1600_udl = og.create_load(
        loadtype="patch",
        name="M1600 Lane UDL",
        point1=vertex_1,
        point2=vertex_2,
        point3=vertex_3,
        point4=vertex_4,
    )
    m1600_lc_list[i].add_load(load_obj=M1600_udl,load_factor=alf[i])
    model.add_load_case(m1600_lc_list[i],load_factor=dla)
```

Finally, we add the defined M1600 lane loads to a load case named `M1600_lc`, then adding it to `model` and applying the `dla` to the load case.

### Moving load case

We can make the M1600 into a [moving load case](https://monashsmartstructures.github.io/ospgrillage/rst/Running_analysis.html#moving-load) instead. 

For this example, lets move the M1600 starting from its initial (static load case) position to 
4 meters in the x direction.



#### Create moving path

First we create a `Path` object for the load model. A [path](https://monashsmartstructures.github.io/ospgrillage/_images/movingload.png) is defined using two `Point` named tuple defined either by `create_point()` or directly `Point(x,y,z)`.

Since the origin of the truck compound load coordinate system is at the back axle $x_{local} = 0$ m, we need to place the start point at $x_{global}=-25 $ m and the end point at the end of the bridge.

```python
start = og.create_point(x=-25,y=0,z=0)
end = og.Point(L,0,0)
m1600_path = og.create_moving_path(start_point=start, end_point= end)
```

#### Create moving load 

Create moving load object; adding Path and Load components to moving load.

We do this using a `for` loop again. Since the load components of the M1600 have been previously created and stored in `M1600_moving_loads`, we can just add it to the moving load.

```python
moving_load_list = []
for i in range(3):
    name = f"Moving M1600 L{i+1}"
    moving_m1600 = og.create_moving_load(name=name) 
    moving_m1600.set_path(m1600_path)
    moving_m1600.add_load(M1600_moving_loads[i])
    model.add_load_case(moving_m1600)
    #add to list for later on
    moving_load_list.append(moving_m1600)
```

The moving load is parsed into load cases for each incremental position in the path. The default increment is 50.

```python
# set beams
shell_model.set_member(longitudinal_beam, member="interior_main_beam")
shell_model.set_member(longitudinal_beam, member="exterior_main_beam_1")
shell_model.set_member(longitudinal_beam, member="exterior_main_beam_2")
# set shell
shell_model.set_shell_members(slab_shell)
```
